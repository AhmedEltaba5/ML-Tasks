```python
categorical_cols = [col for col in df.columns if df[col].dtype == 'object']
numerical_cols = [col for col in df.columns if df[col].dtype in ['int64', 'float64']]
categorical_cols, numerical_cols
```

## function to Count number of words in categorical feature

```python
def count_words(text):
    text = text.strip()
    if text == "":
        return 0
    return len(text.split(' '))
```

```python
text = df.iloc[0,0] 
count_words(text)
```

```python
for col in categorical_cols:
    count_col = col + "_n_words"
    df[count_col] = df[col].apply(lambda x: count_words(x))
```

```python
df.head()
```

```python
def plot_word_count(col):
    df[col + "_n_words"].value_counts().plot(kind='bar')
    plt.xlabel(f'number of words in {col}')
    plt.title(f'distribution of number of words in {col}')
    plt.show()
```

```python
for col in categorical_cols:
    plot_word_count(col)
```

## Get Top 10

```python
def get_top_10(col):
    publisher_plot = sns.barplot([get_display( arabic_reshaper.reshape(i)) for i in df[col].value_counts()[0:10].index],\
                             df[col].value_counts()[0:10])

    for item in publisher_plot.get_xticklabels():
        item.set_rotation(90)
    plt.title(f'Top 10 {col}')
    plt.show()
```

### There are repeated covers

```python
def covertype(x):
    x = str(x).strip().lower()
    if x=='Hard Cover' or x=='HardCover' or x=='hardcover' or x=='hard cover' or x== 'مقوى':
        return 'hardcover'
    if x=='Paperback' or x=='paperback' or x=='غلاف ورقي' or x=='غلاف ورقي':
        return 'paperback'
    if x=='عادي' or x=='غلاف عادي' or x=='غلاف':
        return 'standard'
    return x
```

```python
df.Cover = df.Cover.apply(lambda x: covertype(x))
```
